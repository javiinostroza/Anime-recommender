# Scrapping de usuarios

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Primero cargamos todos los datos como DataFrames, para poder manejarlos mejor.

Male users:

```python
male_users = pd.read_csv('scrapping_anime_list_by_user/male_users.csv', encoding='utf-8', names = ['user', 'gender']) # This is a DataFrame
male_users = male_users.drop_duplicates()
male_users["gender"] = male_users["gender"].replace(["None"], "Male")
male_users
```

Female users:

```python
female_users = pd.read_csv('scrapping_anime_list_by_user/female_users.csv', encoding='utf-8', names = ['user', 'gender']) # This is a DataFrame
female_users = female_users.drop_duplicates()
female_users["gender"] = female_users["gender"].replace(["None"], "Female")
female_users
```

Non-binary users

```python
#non_bin_users = pd.read_csv('scrapping_anime_list_by_user/non_bin_users.csv', encoding='utf-8', names = ['user', 'gender']) # This is a DataFrame
#non_bin_users = non_bin_users.drop_duplicates()
#non_bin_users["gender"] = non_bin_users["gender"].replace(["None"], "Non-Binary")
#non_bin_users
```

### Con los datos de los usuarios cargados, ahora formamos un solo DataFrame

Data con todos los usuarios: 

```python
users = female_users.append([male_users])
users = users.drop(users[users.gender == "Not Found"].index)
users # DataFrame final de todos los usuarios
```

Usuarios según género

```python
users.gender.value_counts()
```

```python
fig, ax = plt.subplots()
plt.bar(users.gender.value_counts().index, users.gender.value_counts())
plt.title('Géneros de usuarios')             
ax.set_xlabel('Cantidad')           
ax.set_ylabel('Cantidad de usuarios según su género') 
plt.show()
```

# Ahora, convertimos los ratings en DataFrames

Female ratings:

```python
female_ratings = pd.read_csv('scrapping_anime_list_by_user/female_ratings.csv', encoding='utf-8', names = ['user', "anime_id", 'anime', 'rating'])
female_ratings
```

Male ratings:

```python
male_ratings = pd.read_csv('scrapping_anime_list_by_user/male_ratings.csv', encoding='utf-8', names = ['user', "anime_id",'anime', 'rating'])
male_ratings
```

```python
#Non binary ratings:
#non_bin_ratings_1 = pd.read_csv('ratings_csv/non_bin_ratings_1.csv', encoding='utf-8', names = ['user', 'anime', 'rating'])
#non_bin_ratings_1
```

**Data con todos los ratings:**

```python
ratings = female_ratings.append([male_ratings])
ratings = ratings.drop(["anime"], axis = 1)
ratings # DataFrame final de los ratings
```

### Finalmente mezclamos users y ratings para obtener nuestra base de datos final:

```python
result = pd.merge(users, ratings, how='left', on='user')
result = result.dropna()
result = result.drop_duplicates()
result['anime_id'] = result['anime_id'].astype(int) 
result.to_csv("datasets_finales/usuarios.csv", sep = ';', index=False)
result # DataFrame final
```

**Descripción estadística de los ratings de los usuarios:**

```python
result.describe()
```

**Ratings de animes según el género del usuario**

```python
df = result
df = df[df.gender != "None"]
df=df.groupby(['rating','gender']).size()
df=df.unstack()
df.plot(kind='bar')
plt.title('Ratings de animes según el género del usuario') 
plt.show()
```


### Combinamos las bases de datos de animes con la de ratings

```python
animes = pd.read_csv("datasets_finales/animes.csv", sep = ";")
```
